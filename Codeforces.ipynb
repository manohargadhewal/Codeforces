```python
#float or int 
n = input().strip()

if "." in n:
    a, b = n.split(".")
    if int(b) == 0:
        print("int", a)
    else:
        print("float", a, "0." + b)
else:
    print("int", n)
```

```python
 #comparision
A, S, B = input().split()
A = int(A)
B = int(B)

if S == '<':
    print("Right" if A < B else "Wrong")

elif S == '>':
    print("Right" if A > B else "Wrong")

elif S == '=':
    print("Right" if A == B else "Wrong")
```

```python
# Mathsmatic Expression
A, S, B, Q, C = input().split()
A = int(A)
B = int(B)
C = int(C)
if S == '+':
    result = A + B
elif S == '-':
    result = A - B
elif S == '*':
    result = A * B
if result == C:
    print("Yes")
else:
    print(result)
```
